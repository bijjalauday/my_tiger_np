```python
import os

def rename_nyt_vision_txt_files(folder_path):
    for filename in os.listdir(folder_path):
        if filename.endswith('.txt') and 'nyt_vision' in filename:
            old_path = os.path.join(folder_path, filename)
            new_filename = filename.replace('_nyt_vision', '')
            new_filename = f"nyt_vision_{new_filename}"
            new_path = os.path.join(folder_path, new_filename)
            os.rename(old_path, new_path)
            print(f"Renamed {filename} to {new_filename}")

# Replace 'folder_path' with the path to the folder containing the .txt files
# /home/ec2-user/SageMaker/full_data/train/labels/
folder_path = "/home/ec2-user/SageMaker/full_data/train/labels/"
rename_nyt_vision_txt_files(folder_path)
```
