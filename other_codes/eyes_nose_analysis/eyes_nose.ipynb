```python
import pybboxes as pbx
import numpy as np 
import cv2
import pandas as pd 
import os 
miss=[]


for i in os.listdir("D:/kaggle_keypoint_data/archive/CAT_06"):      # Downloaded the data kaggle 
    if "cat" in i: 
        j=i.split(".ca")[0]     #The text files are present in the format security catalog. The file names are like 00001292_000.jpg.cat 
                                # So we car checking whether a file name has ca.
                                # Before ca, it will be the jpg file name. So, in i, we have security catalog label file and in j we will be having jpg file 
        df=pd.read_csv(f"D:/kaggle_keypoint_data/archive/CAT_06/{i}",header=None,sep=" ")
        img=cv2.imread(f"D:/kaggle_keypoint_data/archive/CAT_06/{j}")
        width_scale=256/img.shape[1]
        height_scale=256/img.shape[0]
        img=cv2.resize(img,(256,256))   #Resizing the images into 256,256.  
        xl=df.iloc[0].values[1]*width_scale     # Changing the values of left eye and right eye midpoints to 256 scale
        yl=df.iloc[0].values[2]*height_scale
        xr=df.iloc[0].values[3]*width_scale
        yr=df.iloc[0].values[4]*height_scale
        xmid=(xl+xr)/2                    # Took the midpoint of left eye mid point and right eye midpoint
        xlmax=(xmid+xl)/2                 # Took the mid point of above point and
        xlmin=xl-(xlmax-xl)
        xrmin=(xmid+xr)/2
        xrmax=xr+abs((xr-xrmin))
        xmou=df.iloc[0].values[5]*width_scale
        ymou=df.iloc[0].values[6]*height_scale
        lymid=(yl+ymou)/2
        lymid2=(yl+lymid)/2
        ylmax=yl+(lymid2-yl)
        ylmin=yl-(lymid2-yl)
        # rymid=(yr+ymou)/2
        # rymid2=(yr+rymid)/2
        # yrmax=rymid2
        # yrmin=yl-(rymid2-yr)
        yrmax=yr+(ylmax-yl)
        yrmin=yr-(yl-ylmin)
        # mx=103*width_scale
        # my=146*height_scale 
        ymoumin=ymou-(yl-ylmin)
        ymoumax=ymou+(yl-ylmin)
        xmoumin=xmou-(xlmax-xl)
        xmoumax=xmou+(xlmax-xl) 
        # cv2.rectangle(img, (int(xlmin), int(ylmin)), (int(xlmax),int(ylmax)), (0, 255, 0), 1)
        # cv2.rectangle(img, (int(xrmin), int(yrmin)), (int(xrmax),int(yrmax)), (0, 255, 0), 1)
        # cv2.rectangle(img, (int(xmoumin), int(ymoumin)), (int(xmoumax),int(ymoumax)), (0, 255, 0), 1)
        try:
            voc_bbox = (xlmin, ylmin, xlmax, ylmax)
            l=pbx.convert_bbox(voc_bbox, from_type="voc", to_type="yolo", image_size=(256,256))
            xlmin1,ylmin1,xlmax1,ylmax1=l[0],l[1],l[2],l[3]
            voc_bbox=(xrmin,yrmin,xrmax,yrmax)
            l=pbx.convert_bbox(voc_bbox, from_type="voc", to_type="yolo", image_size=(256,256))
            xrmin1,yrmin1,xrmax1,yrmax1=l[0],l[1],l[2],l[3]
            voc_bbox=(xmoumin,ymoumin,xmoumax,ymoumax)
            l=pbx.convert_bbox(voc_bbox, from_type="voc", to_type="yolo", image_size=(256,256))
            xmoumin1,ymoumin1,xmoumax1,ymoumax1=l[0],l[1],l[2],l[3]

            with open(f"D:/kaggle_keypoint_data/bb_labels/{j}.txt","w") as file:
                file.write(str(0))
                file.write(" ")
                file.write(str(np.round(xlmin1,decimals=3)))
                file.write(" ")
                file.write(str(np.round(ylmin1,decimals=3)))
                file.write(" ")
                file.write(str(np.round(xlmax1,decimals=3)))
                file.write(" ")
                file.write(str(np.round(ylmax1,decimals=3)))
                file.write("\n")
                file.write(str(0))
                file.write(" ")
                file.write(str(np.round(xrmin1,decimals=3)))
                file.write(" ")
                file.write(str(np.round(yrmin1,decimals=3)))
                file.write(" ")
                file.write(str(np.round(xrmax1,decimals=3)))
                file.write(" ")
                file.write(str(np.round(yrmax1,decimals=3)))
                file.write("\n")
                file.write(str(1))
                file.write(" ")
                file.write(str(np.round(xmoumin1,decimals=3)))
                file.write(" ")
                file.write(str(np.round(ymoumin1,decimals=3)))
                file.write(" ")
                file.write(str(np.round(xmoumax1,decimals=3)))
                file.write(" ")
                file.write(str(np.round(ymoumax1,decimals=3)))

        except:
            print(f"cant do for this {i}")
            miss.append(i)

        
        # cv2.imwrite(f"D:/kaggle_keypoint_data/bb_images/{j}",img)
        
```
