```python
import pandas as pd 
l1=[]
for i in os.listdir("D:/cat_id/yolov5/runs/detect/exp4/labels/"):
    l2=[]
    l2.append(i)
    if "side" in i:
        orig="side"
    else:
        orig="front"
    l=pd.read_csv(f"D:/cat_id/yolov5/runs/detect/exp4/labels/{i}",header=None,sep=" ")
    c=0
    for i in l.iterrows():
        if i[1][0]==0:
            c=c+1
    if c==2:
        pred="front"
    elif c==1:
        pred="side"
    else:
        pred=-1
    l2.append(orig)
    l2.append(pred)
    l1.append(l2)
```

```python
df=pd.DataFrame(l1,columns=["name","orig","pred"])
```

```python
df1=df[df["pred"]!=-1]
from sklearn.metrics import confusion_matrix,classification_report 

print(classification_report(df1["orig"],df1["pred"]))
print(confusion_matrix(df1["orig"],df1["pred"]))
```
