```python
l2=[]
for i in os.listdir("D:/cat_id/deep_dive/1611/OneDrive_2023-11-14-PTC_approach_11-9-2023/PTC_approach_11-9-2023"):
    j=i.split(".mp")[0]
    cap = cv2.VideoCapture(f"D:/cat_id/deep_dive/1611/OneDrive_2023-11-14-PTC_approach_11-9-2023/PTC_approach_11-9-2023/{i}")
    c=0
    while True:
        # Read a frame from the video
        ret, frame = cap.read()

        # Break the loop if we have reached the end of the video
        if not ret:
            break
        c=c+1
    l=[j,c]
    l2.append(l)
```

```python
import pandas as pd
l1=[]
for i in os.listdir("./other"):
    if "analysis" in i:
        continue
    ca=0
    do=0
    ca91=0
    ca89=0
    ca78=0
    ca67=0
    ca6l=0
    do91=0
    do89=0
    do78=0
    do67=0
    do6l=0
    multiple=0
    for j in os.listdir(f"other/{i}/labels"):
        l=pd.read_csv(f"other/{i}/labels/{j}",header=None,sep=" ")
        if len(l)==1:
            if l.iloc[0,:].values[0]==0:
                ca=ca+1
                cs=l.iloc[0,:].values[-1]
                if cs>=0.9:
                    ca91=ca91+1
                elif cs>=0.8:
                    ca89=ca89+1
                elif cs>=0.7:
                    ca78=ca78+1
                elif cs>=0.6:
                    ca67=ca67+1
                else:
                    ca6l=ca6l+1
            if l.iloc[0,:].values[0]==1:
                do=do+1
                cs=l.iloc[0,:].values[-1]
                if cs>=0.9:
                    do91=do91+1
                elif cs>=0.8:
                    do89=do89+1
                elif cs>=0.7:
                    do78=do78+1
                elif cs>=0.6:
                    do67=do67+1
                else:
                    do6l=do6l+1
        else:
            multiple=multiple+1
    l=[i,ca,do,ca91,ca89,ca78,ca67,ca6l,do91,do89,do78,do67,do6l,multiple]
    l1.append(l)
            
```

```python
df=pd.DataFrame(l1,columns=["video_name","# cat frames","# dog frames","#cat frames_0.9-1.0_conf","#cat frames_0.8-0.9_conf","#cat frames_0.7-0.8_conf","#cat frames_0.6-0.7_conf","#cat frames_less_than_0.6","#dog frames_0.9-1.0_conf","#dog frames_0.8-0.9_conf","#dog frames_0.7-0.8_conf","#dog frames_0.6-0.7_conf","#dog frames_less_than_0.6","multiple"])
```
